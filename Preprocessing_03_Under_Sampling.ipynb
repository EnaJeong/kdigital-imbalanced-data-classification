```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from collections import Counter
from sklearn.datasets import make_classification
from matplotlib import pyplot
from numpy import where
import imblearn
from imblearn.under_sampling import RandomUnderSampler
from imblearn.under_sampling import TomekLinks
from imblearn.under_sampling import CondensedNearestNeighbour
from imblearn.under_sampling import OneSidedSelection
from imblearn.under_sampling import EditedNearestNeighbours
from imblearn.under_sampling import NeighbourhoodCleaningRule
```

## 샘플 데이터 세팅 

```python
# 표본 데이터 수
# 독립 변수의 수
# 독립 변수 중 다른 독립 변수의 선형 조합으로 나타나는 성분의 수
# 클래스 당 클러스터의 수 
# 각 클래스의 할당된 표본 수
# 클래스가 임의로 교환되는 샘플의 일부, 라벨에 노이즈를 생성해 분류를 어렵게 만듬
# random_state = 22
X, y = make_classification(n_samples=1000, 
                           n_features = 2, 
                           n_redundant=0,
                          n_clusters_per_class = 1, 
                           weights=[0.98],
                          flip_y = 0, 
                           random_state = 22)
```

```python
# 데이터셋 정의
X, y = make_classification(n_samples=1000, n_features=2, n_redundant=0,
    n_clusters_per_class=1, weights=[0.98], flip_y=0, random_state=22)
# 클래스 별 분포 요약
counter = Counter(y)
print(counter)
# 스캐터 플롯 그리기 
for label, _ in counter.items():
    row_ix = where(y == label)[0]
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label), alpha = 0.5)
pyplot.legend()
pyplot.show()
```

## RandomUnderSampler

```python
# 데이터셋 정의
X, y = make_classification(n_samples=1000, n_features=2, n_redundant=0,
    n_clusters_per_class=1, weights=[0.98], flip_y=0, random_state=22)

# Sampling
rus = RandomUnderSampler()
X, y = rus.fit_resample(X,y)

# 클래스 별 분포 요약
counter = Counter(y)
print(counter)
# 스캐터 플롯 그리기 
for label, _ in counter.items():
    row_ix = where(y == label)[0]
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label), alpha = 0.5)
pyplot.legend()
pyplot.show()
```
