```python
fig, axes = plt.subplots(2, 2, figsize=(20, 10))

key_list = ['Amount', 'Time']

for ax, key in zip(axes, key_list):
    data = df[key]
    
    sns.distplot(data, ax=ax[0])
    ax[0].set_xlim([min(data), max(data)])
    
    sns.boxplot(x=key, data=df, ax=ax[1])

plt.show()
```

## Scaling

```python
from sklearn.preprocessing import MinMaxScaler
# from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
```

```python
min_max_scaler = MinMaxScaler()
# max_abs_scaler = MaxAbsScaler()
standard_scaler = StandardScaler()
robust_scaler = RobustScaler()
```

```python
target = 'Amount'

data = df[target].values.reshape(-1, 1)

scaled_by_min_max = min_max_scaler.fit_transform(data)
scaled_by_standard = standard_scaler.fit_transform(data)
scaled_by_robust = robust_scaler.fit_transform(data)

print(f"Raw      : min={min(data)[0]}, max={max(data)[0]}")
print(f"Min Max  : min={min(scaled_by_min_max)[0]}, max={max(scaled_by_min_max)}[0]")
print(f"Standard : min={min(scaled_by_standard)[0]}, max={max(scaled_by_standard)}[0]")
print(f"Robust   : min={min(scaled_by_robust)[0]}, max={max(scaled_by_robust)}[0]")


# 그래프
fig, axes = plt.subplots(4, 1, figsize=(10, 16))

data_list = {'Raw': df[target].values, 'Min Max': scaled_by_min_max, 
             'Standard': scaled_by_standard, 'Robust': scaled_by_robust}
    
for ax, (key, value) in zip(axes.ravel(), data_list.items()):
    sns.boxplot(x=value, data=df, ax=ax)
    ax.set_title(key)

plt.show()
```
